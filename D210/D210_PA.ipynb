### C5: Accessibility for Colorblindness


- In designing the dashboard, colorblindness accessibility is taken into account. This goal was achieved by selecting colors that are distinguishable to individuals with color vision deficiencies. Strict avoidance of red-green gradients ensured that those with specific problems with those colors would be accounted for properly. Light and dark shades were also used to help with differentiation, and blue-orange gradients were due to their safeness for most color-blind issues. Contrasting colors ensured the differences were substantial enough to visualize different data and trends, regardless of the colors used.
